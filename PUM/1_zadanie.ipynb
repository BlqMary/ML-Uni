```python
import random
import numpy as np
import math 
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (16,8)
```

### Zadanie 1

```python
def getPointsInsideHypercirclePercent(dim, n, r = 1):
    H = np.array([[random.random() * 2 - 1 for y in range(dim)] for x in range(n)])
    countIn = 0
    rSquared = pow(r,2)
    for j in range(n): 
        sumDist = sum(pow(H[j],2))
        if(sumDist <= rSquared):
            countIn += 1

    return countIn/n * 100
```

```python
d = range(1,10)
n = 100
points = 1000

resultsA = [[0 for y in range(n)] for x in d]
for i in d:
    for j in range (n):
        resultsA[i-1][j] = getPointsInsideHypercirclePercent(i, points)
    print("Dimention:", i, "done")
```

```python
test = resultsA
```

```python
avgInForDim = []
avgOutForDim = []
stdForDim = []
for i in d:
    npTest = np.array(test[i-1])
    avgInForDim.append(np.mean(npTest))
    avgOutForDim.append(100 - avgInForDim[i-1])
    stdForDim.append(np.std(npTest))
    print(f'Dimentions: {i} | Inside: {avgInForDim[i-1]:.2f}% | Outside: {avgOutForDim[i-1]:.2f}% | Standard Deviation: {stdForDim[i-1]:.2f} pp')
```

```python
fig, ax = plt.subplots()
ax.boxplot(test)
plt.xlabel("Ilość wymiarów")
plt.ylabel("% punktów w środku")
plt.show()
```

```python
plt.bar(d, avgInForDim, color='bisque', ecolor='dimgray', yerr=stdForDim)
plt.xlabel("Ilość wymiarów")
plt.ylabel("% punktów w środku")
plt.grid(True)
```

### Zadanie 2

```python
def getDistancesBetweenPoints(dim, n, r = 1):
    H2 = np.array([[random.random() * 2 - 1 for y in range(dim)] for x in range(n)])
    distances = []
    
    for i in range(n): 
        for j in range(i+1,n): 
            dist = math.sqrt(sum(pow(H2[i]-H2[j],2)))
            distances.append(dist)
            
    npDistances = np.array(distances)
    return (np.mean(npDistances),np.std(npDistances))
```

```python
d = range(1,10)
n = 100
points = 100

resultsB = [[_ for y in range(n)] for x in d]
for i in d:
    for j in range (n):
        resultsB[i-1][j] = getDistancesBetweenPoints(i, points)
    print("Dimention:", i, "done")
```

```python
test = resultsB
```

```python
l1 = []
l2 = []
l3 = []
l4 = []

for i in d:
    npTest = np.array(test[i-1])
    l1.append(np.mean(npTest[0]))
    l2.append(np.std(npTest[0]))
    l3.append(np.mean(npTest[1]))
    l4.append(np.std(npTest[1]))

    print(f'Dimentions: {i} | Mean Dist: {l1[i-1]:.2f} | Mean Dist StD: {l2[i-1]:.2f} | Mean of StD : {l3[i-1]:.2f} | StD of StD: {l4[i-1]:.2f} | Mean StD/Mean Distance : {l3[i-1]/l1[i-1]*100:.2f}%')
```

```python
plt.errorbar(d, l1, l2, fmt='o', color='darkorange', ecolor='dimgray', marker='s')
plt.xlabel("Ilość wymiarów")
plt.ylabel("Średnia odległość punktów")
plt.ylim(0,2.7)
```

```python
plt.errorbar(d, l3, l4, fmt='o', color='darkorange', ecolor='dimgray', marker='s')
plt.xlabel("Ilość wymiarów")
plt.ylabel("Średnie odchylenie standardowe odległości punktów")
plt.ylim(0,2.7)
```

```python
l = np.array(l3)/np.array(l1) * 100
plt.plot(d,l,'o', color='darkorange', marker='s')
plt.xlabel("Ilość wymiarów")
plt.ylabel("% odchylenia standardowego w średniej")
plt.grid(True)
```

### Zadanie 3

```python
def getVector(p1, p2):
    return p1 - p2

def getAngle(vector1, vector2):
    unit_vector1 = vector1 / np.linalg.norm(vector1)
    unit_vector2 = vector2 / np.linalg.norm(vector2)
    dot_product = np.dot(unit_vector1, unit_vector2)
    return np.absolute(dot_product)

def pop(array):
    index = random.randint(0, len(array) - 1)
    item = array[index]
    new_array = np.delete(array, 1, 0)
    return new_array, item

def getAngleBetweenPoints(dim, n):
    H3 = np.array([[random.random() * 2 - 1 for y in range(dim)] for x in range(n)])

    H3, p1 = pop(H3)
    H3, p2 = pop(H3)
    H3, p3 = pop(H3)
    H3, p4 = pop(H3)

    vector1 = getVector(p1, p2)
    vector2 = getVector(p3, p4)
    angle = getAngle(vector1, vector2)
    
    return angle
```

```python
d = range(2,31)
n = 500
points = 1000

resultsC = [[] for x in d]
for i in d:
    for j in range (n):
        angle = getAngleBetweenPoints(i, points)
        if not np.isnan(angle):
            resultsC[i-2].append(angle)
    print("Dimention:", i, "done")
```

```python
test = resultsC
```

```python
l1 = []
l2 = []

for i in d:
    npTest = np.array(test[i-2])
    l1.append(np.mean(npTest))
    l2.append(np.std(npTest))

    print(f'Dimentions: {i} | Mean Angle: {l1[i-2]:.2f} | Standard Deviation: {l2[i-2]:.2f} ')
```

```python
plt.errorbar(d, l1, l2, fmt='o', color='darkorange', ecolor='dimgray', marker='s')
plt.xlabel("Ilość wymiarów")
plt.ylabel("Średni kąt")
plt.ylim(0,1)
```
